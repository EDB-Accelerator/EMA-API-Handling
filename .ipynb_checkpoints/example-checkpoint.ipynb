```python
import pandas as pd, pytz          # pip install pytz if you don‚Äôt have it

# df  ‚Üê‚Äì DataFrame returned by get_schedule()
# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# 1) Identify any timestamp-like columns that came back as strings
ts_cols = [
    c for c in df.columns
    if c in ("scheduledTime", "startTime", "timeStart", "timeEnd")  # add more if needed
       and df[c].notna().any()
]

# 2) Parse those strings ‚Üí timezone-aware pandas Timestamps
eastern = pytz.timezone("US/Eastern")            # same zone get_schedule() used
for c in ts_cols:
    df[c] = pd.to_datetime(df[c]).dt.tz_localize(eastern)

# 3) Build a mask for ‚Äúany timestamp in the future‚Äù
now = pd.Timestamp.now(tz=eastern)
future_mask = pd.Series(False, index=df.index)
for c in ts_cols:
    future_mask |= df[c] > now

# 4) Slice the frame
df_future = df[future_mask].copy().reset_index(drop=True)

print(f"Existing future beeps: {len(df_future)}")
display(df_future)

```

```python
import json, time, pathlib, pandas as pd, datetime as _dt, ast
from collections import OrderedDict, abc as _abc

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ config
ORDER = [
    "itemId", "localId", "endTime", "startTime",
    "beepId", "scheduledTime", "reminderIntervals", "randomizationScheme"
]
TIME_KEYS = {"startTime", "endTime", "scheduledTime"}

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ helpers
def fmt_time(v):
    """Return 'YYYY-MM-DD HH:MM:SS' (strip T/offset)."""
    if pd.isna(v):
        return None
    if isinstance(v, (pd.Timestamp, _dt.datetime)):
        return v.strftime("%Y-%m-%d %H:%M:%S")
    return v.replace("T", " ").split("+")[0]

def fix_reminders(v):
    """Return list[int] or None."""
    if v is None or (isinstance(v, (list, tuple)) and len(v) == 0) or pd.isna(v):
        return None
    if isinstance(v, str):
        try:
            parsed = ast.literal_eval(v.strip())
            if isinstance(parsed, _abc.Sequence):
                return list(parsed)
        except Exception:
            return None
    if isinstance(v, _abc.Sequence):
        return list(v)
    return None

def to_api_dict(src: dict | pd.Series, idx: int, is_new: bool) -> OrderedDict:
    """Build an OrderedDict in desired key order."""
    od = OrderedDict()
    for k in ORDER:
        if k == "beepId":
            # od[k] = 0 if is_new else src.get(k)          # 0 for new, keep old
            od[k] = 0
        elif k == "localId":
            od[k] = src.get(k) or f"auto_{idx+1:04d}"    # generate if missing
        elif k == "reminderIntervals":
            ri = fix_reminders(src.get(k))
            if ri:
                od[k] = ri
        elif k in src and pd.notna(src[k]):
            od[k] = fmt_time(src[k]) if k in TIME_KEYS else src[k]
    return od

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ build combined
future_entries = [
    to_api_dict(row, i, is_new=False)
    for i, (_, row) in enumerate(df_future.iterrows())
]
new_entries = [
    to_api_dict(e, i + len(future_entries), is_new=True)
    for i, e in enumerate(new_beeps)
]

combined = future_entries + new_entries
print(f"Final list: {len(combined)} beeps "
      f"({sum('localId' in e for e in combined)} with localId)")

# ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ save JSON
ts = time.strftime("%Y%m%dT%H%M%S")
json_path = pathlib.Path(f"upload_ready_{ts}.json").resolve()
json_path.write_text(json.dumps(combined, ensure_ascii=False, indent=2))
print("‚úì Upload-ready JSON ‚Üí", json_path)
display(combined)
```
